**Q.2** Fourth root of a number

```
# METHOD I
import math

def fourth_root(num):
    return math.sqrt(math.sqrt(num))
```

```
num = float(input())
print("Fourth root of {} is {}".format(num, fourth_root(num)))
```

```
# METHOD II

def fourth_root_ii(num):
    return num ** 0.25  # 1 / 4  = 0.25
print("Fourth root of {} is {}".format(num, fourth_root_ii(num)))
```

**Q.3** SWAP TWO NUMBERS using "temp"

```
a = int(input())
b = int(input())
print("First number is {} and second number is {}".format(a, b))
temp = a
a = b
b = temp
print("First number is {} and second number is {}".format(a, b))
```

**Q.4** SWAP NUMBER without using "temp"

```
a = int(input())
b = int(input())
print("First number is {} and second number is {}".format(a, b))
a, b = b, a
print("First number is {} and second number is {}".format(a, b))
```

**Q.5** FAHRENHEIT TO KELVIN & CELSIUS

```
def fahrenheit(val):
    return ((val - 32) * (5 / 9) + 273.15)
def celsius(val):
    return ((val - 32) * (5 / 9))
```

```
val = int(input())
print("{}° farhenheit is {}° kelvin".format(val, fahrenheit(val)))
print("{}° farhenheit is {}° celsius".format(val, celsius(val)))
```
